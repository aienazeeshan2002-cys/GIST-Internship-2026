```python
import re

def analyze_phishing_vector(sender_email: str, message_body: str) -> dict:
    """
    Processes email metadata and string layout to identify common social
    engineering indicators, domain mismatches, and cognitive triggers.
    """
    red_flags = []
    suspicious_keywords_found = []

    # 1. Input Validation Check
    if not sender_email or not message_body:
        return {"status": "ERROR", "message": "Sender domain and email payload cannot be empty."}

    # 2. Vector Check: Cognitive Triggers & Urgency Keywords (Fixed Syntax)
    urgency_patterns = [
        r"urgent", r"action required", r"suspended",
        r"verify your account", r"unauthorized login",
        r"wire transfer", r"password reset", r"immediate"
    ]

    for pattern in urgency_patterns:
        if re.search(pattern, message_body.lower()):
            suspicious_keywords_found.append(pattern)

    if suspicious_keywords_found:
        red_flags.append(f"Psychological Urgency/Pretext triggers detected: {', '.join(suspicious_keywords_found)}")

    # 3. Vector Check: Technical Domain Spoofing Detection
    common_spoofs = ["paypal", "amazon", "google", "facebook", "bank", "secure", "verify"]
    sender_lower = sender_email.lower()

    for brand in common_spoofs:
        if brand in sender_lower and not any(sender_lower.endswith(valid) for valid in [f"@{brand}.com", f".{brand}.com"]):
            red_flags.append(f"Potential Lookalike Domain Spoofing detected! From address contains '{brand}' but uses an unverified domain layout.")

    # 4. Vector Check: Structural Mismatches (e.g., Generic Greetings in Mass Lures)
    generic_greetings = ["dear customer", "dear user", "valued client", "attention account holder"]
    if any(greet in message_body.lower() for greet in generic_greetings):
        red_flags.append("Generic greeting used. Indicates high-volume mass phishing delivery rather than verified business contact.")

    # 5. Risk Assessment Logic
    flag_count = len(red_flags)
    if flag_count >= 3:
        risk_level = "CRITICAL THREAT (Immediate Quarantine Recommended)"
        is_unsafe = True
    elif flag_count >= 1:
        risk_level = "SUSPICIOUS (Further Investigation Required)"
        is_unsafe = True
    else:
        risk_level = "LOW RISK (No Common Vector Triggers Identified)"
        is_unsafe = False

    return {
        "status": "SUCCESS",
        "risk_level": risk_level,
        "is_unsafe": is_unsafe,
        "red_flags": red_flags if red_flags else ["No immediate anomalies identified."],
        "metrics": f"{flag_count} Technical Red Flag Indicators Identified."
    }


def main():
    print("=" * 65)
    print("      DECODELABS SECURITY OPERATION CENTER: PHISHING TRIAGE      ")
    print("=" * 65)

    # --- PHASE 1: INPUT PHASE (The Bait Capture) ---
    print("\n[+] Enter Inbound Threat Parameters for Investigation:")
    sender = input("Sender From Address (e.g., safety@paypal-secure-login.com): ").strip()
    print("Paste the complete Message Body below (Press Enter then Ctrl+D when finished):")

    # Catch multi-line message bodies cleanly
    lines = []
    while True:
        try:
            line = input()
            lines.append(line)
        except EOFError:
            break
    message_payload = "\n".join(lines)

    if not message_payload:
        print("\n[!] Error: No message body payload detected for analysis.")
        return

    # --- PHASE 2: PROCESS PHASE (Analytical Parsing Engine) ---
    report = analyze_phishing_vector(sender, message_payload)

    if report["status"] == "ERROR":
        print(f"\n[!] Triage Error: {report['message']}")
        return

    # --- PHASE 3: OUTPUT PHASE (The Defensive Assessment) ---
    print(f"\n" + "=" * 25 + " INCIDENT ANALYTICAL REPORT " + "=" * 25)
    print(f"[▶] Investigated Sender : {sender}")
    print(f"[▶] Risk Status Category: ** {report['risk_level']} **")
    print(f"[▶] Metric Evaluation   : {report['metrics']}")

    print("\n[🛡️] Identified Red Flag Indicators:")
    for flag in report["red_flags"]:
        print(f" - {flag}")

    print("\n[💡] Security Logic Explanation:")
    if report["is_unsafe"]:
        print(" -> This message is flagged UNSAFE because it leverages malicious social engineering mechanics.")
        print(" -> It displays specific psychological urgency patterns designed to trick human behavior into bypassing structural network perimeters.")
    else:
        print(" -> No definitive indicators found, but remind the user to never click untrusted or unverified tracking hyperlinks.")
    print("=" * 65)

if __name__ == "__main__":
    main()
```
