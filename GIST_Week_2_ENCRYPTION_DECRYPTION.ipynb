```python
def caesar_cipher(text: str, shift: int, mode: str = 'encrypt') -> str:
    """
    Handles mathematical cryptographic transformation using the Caesar Cipher mechanism.
    Shifts integers based on ASCII mapping and wraps using modular arithmetic % 26.
    """
    # Adjust shift direction based on mode (reversible process)
    if mode == 'decrypt':
        shift = -shift

    transformed_chars = []

    for char in text:
        if char.isalpha():
            # Establish ASCII base offset for uppercase (65) vs lowercase (97)
            ascii_offset = 65 if char.isupper() else 97

            # Step 1: Convert char to integer (0-25 relative position)
            char_pos = ord(char) - ascii_offset

            # Step 2: Apply the shift key and apply the finite wrap rule: (x + n) % 26
            new_pos = (char_pos + shift) % 26

            # Step 3: Map back to ASCII character string array
            transformed_chars.append(chr(new_pos + ascii_offset))
        else:
            # Maintain structural integrity for spaces/punctuation without shifting
            transformed_chars.append(char)

    return "".join(transformed_chars)


def main():
    print("=" * 60)
    print("      DECODELABS CRYPTOGRAPHIC ENGINE: CAESAR CIPHER      ")
    print("=" * 60)

    # 1. INPUT PHASE (Plaintext Raw Data)
    plaintext = input("Enter the plaintext message to encrypt: ").strip()
    if not plaintext:
        print("[!] Error: Input cannot be empty.")
        return

    try:
        shift_key = int(input("Enter rotation shift key (e.g., 3): "))
    except ValueError:
        print("[!] Error: Shift key must be an integer.")
        return

    # 2. PROCESS PHASE (Mathematical Transformation)
    ciphertext = caesar_cipher(plaintext, shift_key, mode='encrypt')
    decrypted_text = caesar_cipher(ciphertext, shift_key, mode='decrypt')

    # 3. OUTPUT PHASE (Display Data Lifecycle)
    print(f"\n--- DATA LIFECYCLE RESULTS ---")
    print(f"[+] Plaintext (Raw Input) : {plaintext}")
    print(f"[+] Ciphertext (Secured)  : {ciphertext}")
    print(f"[+] Decrypted (Recovered) : {decrypted_text}")
    print("=" * 60)

if __name__ == "__main__":
    main()
```
