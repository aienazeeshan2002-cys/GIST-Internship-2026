```python
import string

def check_password_strength(password: str) -> dict:
    """
    Evaluates the strength of a password using efficient,
    short-circuit string handling and security logic.
    """
    # 1. Zero Point Policy: Length Verification
    # Any password under 8 characters is an immediate safety fail
    length = len(password)
    if length < 8:
        return {
            "score": 0,
            "classification": "FAIL (Immediate Danger)",
            "feedback": ["Password is too short. Must be at least 8 characters to prevent brute-force risks."]
        }

    # 2. Pattern Recognition using C-optimized short-circuit evaluation
    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(char in string.punctuation for char in password)

    # Calculate score metrics
    criteria_met = sum([has_upper, has_lower, has_digit, has_symbol])

    feedback = []
    if not has_upper: feedback.append("Missing an uppercase letter.")
    if not has_lower: feedback.append("Missing a lowercase letter.")
    if not has_digit: feedback.append("Missing a numeric digit.")
    if not has_symbol: feedback.append("Missing a special symbol (e.g., !, @, #, $).")

    # 3. Defensive Logic & Risk Classification System
    # Adjusting classification based on length and criteria entropy
    if criteria_met == 4 and length >= 12:
        classification = "STRONG"
    elif criteria_met >= 3 and length >= 8:
        classification = "MEDIUM"
    else:
        classification = "WEAK"

    return {
        "score": criteria_met,
        "classification": classification,
        "feedback": feedback if feedback else ["Excellent! Password meets all complexity guidelines."]
    }

def main():
    print("=" * 50)
    print(" DECODELABS INDUSTRIAL KIT: PASSWORD STRENGTH CHECKER ")
    print("=" * 50)

    user_input = input("Enter a password to evaluate: ").strip()

    if not user_input:
        print("\n[!] Error: Input cannot be empty.")
        return

    result = check_password_strength(user_input)

    print(f"\n--- ANALYSIS RESULT ---")
    print(f"Password Evaluated : {user_input}")
    print(f"Risk Classification: ** {result['classification']} **")
    print(f"Entropy Score      : {result['score']}/4 Criteria Met")

    if result['feedback']:
        print("\nSecurity Recommendations:")
        for note in result['feedback']:
            print(f" - {note}")
    print("=" * 50)

if __name__ == "__main__":
    main()
```
